# Jupyter Notebook 

## ¿Qué es?:

Es una aplicación que permite crear y compartir documentos los cuales pueden contener textos, ecuaciones e imágenes, junto con código que puede ser ejecutado para apoyar el resto del texto. En este curso usaremos esta aplicación ya que en el marco del presente curso utilizaremos esta aplicación por su versatilidad para compartir código y texto. Además las notebooks permiten:
  * Gran variedad de Kernels de cálculo, incluyendo los lenguajes de programación habitualmente utilizados en investigación (Python, Matlab/Octave, R, Julia, etc.).
  * Computación paralelizada en procesadores y/o clusters.
  * Formateado de documentos con [Markdown](https://en.wikipedia.org/wiki/Markdown).
  * Compartir documentos.
  * El documento obtenido finalmente, por un lado, puede ser generado en formatos diferentes (.html, .pdf .py,...)
  * ...

Más información puede encontrarse en la [documentación de Jupyter](http://jupyter.readthedocs.io/en/latest/index.html).

## 2. Celdas para texto:
Como podéis ver en este cuaderno, podemos tener celdas de código y de texto. Se cambia desde el menú Cell dando a *Markdown cell* arriba o con el *shortcut* o `esc + m` para pasar a celda de texto y `esc + y` para volver a tener una celda de código. Ambas se ejecutan con run. 

Estos son los comandos básicos que nos permitirán formatear el texto de nuestro documento a través de los comandos MarkDown:

 * La concatenación del caracter hash (*#*) nos permite definir títulos principales (*#*), sub-títulos (*##*), etc.

# Heading (*#*)

## Sub-heading (*##*)

### Another deeper heading (*###*)

#### ...

 * Para crear una lista como esta simplemente debe dejarse un espacio respecto al margen izquierdo del texto y utilizar algún símbolo o número.
 * Se puede introducir equaciones en latex $\sqrt(12)$
 * Los párrafos deben separarse por una línea en blanco.
 * Dos espacios al final de la línea se interpretan como un cambio de línea.
 * La concatenación de asteríscos permite dar formato al texto introduciendo palabras en *cursiva* (\*) o **negrita** (\*\*).
 * Para introducir un enlace debemos poner entre corchetes el texto mostrado y a continuación, entre paréntesis, la url del enlace, por ejemplo: [Mas comandos](http://daringfireball.net/projects/markdown/syntax)

## 3. Barra de herramientas

La notebook se va guardando automáticamente (autosaved), pero siempre tenemos que revisarlo. Si pone **unsaved** es que no se ha guardado, por lo que le podemos dar a guardar en la barra de herramientas (botón de guardar).

También podemos añadir una celda nueva (botón **+**)

O cortar, copiar y pegar cualquier celda.

Para ejecutar una celda y pasar a la siguiente usamos run en la barra de herramientas o `shift enter`

Podemos parar la ejecución en cualquier momento. Reiniciar el kernel y ejecutar toda la notebook de principio a fin con los *shortcuts* de la barra de herramientas.

## 4. Instalación de paquetes
Se pueden instalar paquetes de **Python** usando la plataforma de conda desde el navigator / environments,  usando `conda install nombre_paquete` en la línea de comandos o bien usando `pip3 install nombre_paquete` también desde la línea de comandos. 

Desde la notebook también podemos lanzar órdenes usando a la máquina usando `!` antes. Es como si usaramos la línea de comandos y no tenemos que salir de la notebook. Así que para instalar un paquete de python sin salir de la notebook podemos hacerlo con `!pip3 install nombre_paquete`:

## 5. Google Colab

[Colaboratory](https://colab.research.google.com/) es un entorno gratuito de Jupyter Notebook que no requiere configuración y que se ejecuta completamente en la nube.

Colaboratory te permite escribir y ejecutar código, guardar y compartir tus análisis y tener acceso a recursos informáticos, todo de forma gratuita desde el navegador.

### ¿Cuál es la diferencia entre Jupyter y Colaboratory?

Jupyter es el proyecto de código abierto en el que se basa Colaboratory. Colaboratory te permite usar y compartir blocs de notas de Jupyter con otras personas sin tener que descargar, instalar ni ejecutar nada en la computadora, además del navegador.
